```python
# EXEMPLO: arquivo_kml = open('Cultura.kml', 'r') arquivo_kml = open('Saúde.kml', 'r')

# VERSÂO 1.0, versão 2.0 em https://github.com/Lroberto-18/BolsaPAP_ufca/blob/main/kml_extract.py

arquivo_kml = open('Cultura.kml', 'r') #,encoding="utf8"
header =['<kml xmlns','</StyleMap>']
pontos =['<Placemark>', '</Placemark>']
anos = ['<value>2018', '<value>2019','<value>2020','<value>2021']
ignorar = '<description>' 
dados_2018 =[]
dados_2019 =[]
dados_2020 =[]
dados_2021 =[]
dados =[]

#salvar arquivo em vetor
for line in arquivo_kml:
  dados.append(line.strip('\n'))
```

```python
#remover os dados da tag '<description>'
control = True
i=0
while(control):
    linha = dados[i]
    if ignorar in linha:
        del dados[i]
    i+=1
    if(i>=len(dados)):
        control=False
```

```python
opcao = True
t1 = False
t2 = False
i = 0
ref1 =[]
#primeiras linhas Header
while(opcao):
  linha = dados[i]
  if header[0] in linha:
    ref1.append(i)
    t1= True
  if header[1] in linha:
    ref1.append(i)
    t2 = True
  if(t1==True and t2==True):
    opcao = False
  i = i + 1

for j in range(0, ref1[1]+1):
  dados_2018.append(dados[j])
  dados_2019.append(dados[j])
  dados_2020.append(dados[j])
  dados_2021.append(dados[j])
  
```

```python
#buscar por anos
opcao = True
t1 = False
t2 = False
i = 0
while(opcao):
  linha = dados[i]
  if pontos[0] in linha:
    ref3 = i
    t1= True
  if pontos[1] in linha:
    ref4 = i
    t2 = True
  if(t1==True and t2==True):
     for j in range(ref3+1, ref4):
      d = dados[j]
      if anos[0] in d:
        for j in range(ref3, ref4+1):
          dados_2018.append(dados[j])
      if anos[1] in d:
        for j in range(ref3, ref4+1):
          dados_2019.append(dados[j])
      if anos[2] in d:
        for j in range(ref3, ref4+1):
          dados_2020.append(dados[j])
      if anos[3] in d:
        for j in range(ref3, ref4+1):
          dados_2021.append(dados[j])
  if(i==len(dados)-1):
      opcao = False
  i = i + 1

#Fechar a tag
dados_2018.append(dados[len(dados)-2])
dados_2019.append(dados[len(dados)-2])
dados_2020.append(dados[len(dados)-2])
dados_2021.append(dados[len(dados)-2])
dados_2018.append(dados[len(dados)-1])
dados_2019.append(dados[len(dados)-1])
dados_2020.append(dados[len(dados)-1])
dados_2021.append(dados[len(dados)-1])
```

```python
#Exportando arquivos em kml
with open('Cultura_2018.kml', 'w') as kml:
  for linha in dados_2018:
    kml.write('{}\n'.format(linha))
with open('Cultura_2019.kml', 'w') as kml:
  for linha in dados_2019:
    kml.write('{}\n'.format(linha))
with open('Cultura_2020.kml', 'w') as kml:
  for linha in dados_2020:
    kml.write('{}\n'.format(linha))
with open('Cultura_2021.kml', 'w') as kml:
  for linha in dados_2021:
    kml.write('{}\n'.format(linha))
```
